# One Hot encoding:

It creates a binary column for each category of the feature. Each row has a value  of 1 in the column corressponding to its category and 0 in the other columns.

Advantage:
Used for nominal variables

Disadvantage:
It can lead to high dimensionality with many categories.

```python
from sklearn.preprocessing import OneHotEncoder
train = pd.read_csv("train.csv")
ohe = OneHotEncoder(sparse=False, drop ="first")
columns=["nom_0","nom_1","nom_2","nom_3","nom_4","nom_5","nom_6","nom_7","nom_8","nom_9"]
ohe_df = pd.DataFrame(ohe.fit_transform(train[columns]))
ohe_df.columns = ohe.get_feature_names_out(columns)
train = train.drop(columns, axis=1)
train = pd.concat([train, ohe_df], axis=1)
print(train.T)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
le = LabelEncoder()
df_le = train.copy()
columns=["ord_3","ord_4","ord_5"]
for col in columns:
    df_le[col + '_encoded'] = le.fit_transform(df_le[col])
train = train.drop(columns, axis=1)
train = pd.concat([train, df_le], axis=1)
print(train.T)
```

```python
#Ordinal Encoding

from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
df_oe = train.copy()
columns=["ord_0","ord_1","ord_2"]
for col in columns:
    df_oe[col + '_encoded'] = oe.fit_transform(df_oe[[col]])
train = train.drop(columns, axis=1)
train = pd.concat([train, df_oe], axis=1)
print(train.T)
```
